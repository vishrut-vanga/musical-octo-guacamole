```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
all_data = pd.read_csv('all_data.csv')
```

```python
print(all_data.head())
```

```python
#All countries that get graph
print(all_data.Country.unique())
```

```python
#Find values to plot for Chile

chile = all_data[all_data['Country'] == 'Chile']
x_vals = chile['Life expectancy at birth (years)']
y_vals = chile['GDP']

x_vals = np.array(x_vals)
x_vals = x_vals.tolist()
print(x_vals)

#Changing the array into np aray to perform operations on
np.array(y_vals)
y_vals = y_vals / 100000000000
y_vals = y_vals.tolist()
print(y_vals)


#Print length of the dataframe of a specific country
#print(len(chile))


```

```python
#Initializing the subplots
plt.subplot(2, 3, 1)
plt.plot(x_vals, y_vals, "o")
plt.title('Chile')
plt.ylabel('GDP in USD in Hundred Billions')
plt.xlabel('Life Expectancy')


plt.subplot(2, 3, 2)
plt.plot(china_x_vals, china_y_vals, "o")
plt.title('China')
plt.ylabel('GDP in USD in Ten Trillions')
plt.xlabel('Life Expectancy')

#Initializing the subplots
plt.subplot(2, 3, 3)
plt.plot(ger_x_vals, ger_y_vals, "o")
plt.title('Germany')
plt.ylabel('GDP in USD in Ten Trillions')
plt.xlabel('Life Expectancy')


#Initializing the subplots
plt.subplot(2, 3, 4)
plt.plot(mex_x_vals, mex_y_vals, "o")
plt.title('Mexico')
plt.ylabel('GDP in USD in Trillions')
plt.xlabel('Life Expectancy')


#Initializing the subplots
plt.subplot(2, 3, 5)
plt.plot(usa_x_vals, usa_y_vals, "o")
plt.title('United States of America')
plt.ylabel('GDP in USD in Ten Trillions')
plt.xlabel('Life Expectancy')

#Initializing the subplots
plt.subplot(2, 3, 6)
plt.plot(zim_x_vals, zim_y_vals, "o")
plt.title('Zimbabwe')
plt.ylabel('GDP in USD in Ten Billions')
plt.xlabel('Life Expectancy')

plt.subplots_adjust(wspace=2, hspace=1)
plt.show()
```
